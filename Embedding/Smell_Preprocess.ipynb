```python
import csv

# Set the field size limit to a specific value
csv.field_size_limit(10**9)

input_file = 'AST_Data/temp/train-data.csv'
output_file = 'AST_Data/temp/train-data_whitout_numbers.csv'

with open(input_file, 'r') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)  # Read the header

    # Remove the numbers from each row
    cleaned_rows = []
    for row in csv_reader:
        cleaned_row = [
            ''.join([char for char in row[0] if not char.isdigit()]),  # Remove numbers from parsed_smell
            ''.join([char for char in row[1] if not char.isdigit()]),  # Remove numbers from parsed_smell_identifiers
            row[2]  # Keep the message as is
        ]
        cleaned_rows.append(cleaned_row)

    # Save the modified content to the output file
    with open(output_file, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow(header)  # Write the header
        csv_writer.writerows(cleaned_rows)

print(f"The file '{output_file}' has been saved with the numbers removed.")
```

```python
import csv
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle

input_file = 'AST_Data/temp/train-data_whitout_numbers.csv'

# Read the modified file
df = pd.read_csv(input_file)

# Extract the "message" column values
messages = df['message'].values

# Perform label encoding
encoder = LabelEncoder()
encoded_messages = encoder.fit_transform(messages)

# Update the DataFrame with the encoded values
df_encoded = df.copy()
df_encoded['message_encoded'] = encoded_messages

# Save the encoded content to a new CSV file
output_file = 'AST_Data/temp/train-data_encoded.csv'
df_encoded.to_csv(output_file, index=False)

# Save the encoding vocabulary
vocabulary_file = 'AST_Data/message_encoding_vocabulary.pkl'
with open(vocabulary_file, 'wb') as f:
    pickle.dump(encoder.classes_, f)

print(f"The encoded data has been saved to '{output_file}'.")
print(f"The message encoding vocabulary has been saved to '{vocabulary_file}'.")
```
