```python
import csv

input_file = 'input-nanr.csv'
output_file = 'input.csv'

with open(input_file, 'r') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)  # Read the header

    # Remove the numbers from each row while keeping the deleted numbers
    cleaned_rows = []
    for row in csv_reader:
        deleted_numbers = ''.join([char for char in row[0] if char.isdigit()])  # Extract and keep the deleted numbers
        cleaned_row = [
            ''.join([char for char in row[0] if not char.isdigit()])  # Remove numbers from parsed_smell
        ]
        line_number = deleted_numbers[:2]  # Extract the first two digits from the deleted numbers as the line number
        cleaned_rows.append([line_number] + cleaned_row)

    # Save the modified content to the output file
    with open(output_file, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        header_with_deleted_number = ['line_number'] + [header[0]]  # Update the header with only the "parsed_smell" column
        csv_writer.writerow(header_with_deleted_number)  # Write the header with added column for line number
        csv_writer.writerows(cleaned_rows)

print(f"The file '{output_file}' has been saved with the numbers removed while keeping the deleted numbers.")
```
